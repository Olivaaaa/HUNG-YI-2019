```python
import numpy as np
import csv
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#用numpy的append用法，将数据拼接
#测试集的项目数据是九个小时的项目数据，因此取前八个小时pm2.5的数据放到x_data中，第9个小时的数据放到预测y_data中
train_x_data = []
train_y_data = []
# data = csv.reader('resources/train.csv')
filename = 'resources/train.csv'
with open(filename, encoding='utf-8') as file:
    reader = csv.reader(file)
    #把所有的数据转成list
    ls = list(reader)
    for i in range(10, len(ls), 18):
        
        for j in range(3, 19):
            train_x_data.append(ls[i][j:j+8])
            train_y_data.append(ls[i][j+8])
#数据转成array格式
train_x_data = np.array(train_x_data, float)
train_y_data = np.array(train_y_data, float)
#存储数据savetxt
np.savetxt("resources/x_data.csv", train_x_data, fmt='%.1f', delimiter=",")
np.savetxt("resources/y_data.csv", train_y_data, fmt='%.1f', delimiter=",")

#测试数据处理
test_x_data = []
test_y_data = []

filename_2 = 'resources/test.csv'
with open(filename_2, encoding='utf-8') as f:
    reader = csv.reader(f)
    ls = list(reader)
    for i in range(9, len(ls), 18):
        test_x_data.append(ls[i][2:10])
        test_y_data.append(ls[i][10])
test_x_data = np.array(test_x_data, float)
test_y_data = np.array(test_y_data, float)

np.savetxt("resources/x_test_data.csv", test_x_data, fmt='%.1f', delimiter=",")
np.savetxt("resources/y_test_data.csv", test_y_data, fmt='%.1f', delimiter=",")
```

```python
#岭回归是一种正则化方法
# 使用广义交叉验证 拟合得到最优alpha参数 = 898.21
#np.linespace在指定的间隔返回均匀间隔的数字
# regs = linear_model.RidgeCV(np.linspace(1,1000))
# regs.fit(x_data, y_data)
# alpha = regs.alpha_
trainingData = 'resources/'
x_data = []
y_data = []
np.loadtxt(open(trainingData+'x_data.csv'), delimiter=",", skiprows=0)
#导入训练数据
with open(trainingData+'x_data.csv', encoding='utf-8') as f:
    x_data = np.loadtxt(f, delimiter=",")
with open(trainingData+'y_data.csv') as f:
    y_data = np.loadtxt(f, delimiter=",")
    
reg = linear_model.Ridge(alpha=898.21)
reg.fit(x_data, y_data)

test_x_data = []
test_y_data = []
#导入测试数据
with open(trainingData+'x_test_data.csv', encoding='utf-8') as f:
    test_x_data = np.loadtxt(f, delimiter=",")
with open(trainingData+'y_test_data.csv', encoding='utf-8') as f:
    test_y_data = np.loadtxt(f, delimiter=",")
    
#画出训练集误差图
pre_y_data = reg.predict(x_data)

err = abs(pre_y_data - y_data)

plt.plot(np.linspace(1, len(err), len(err)), err)
plt.xlabel(r'$trainingIndex$', fontsize=16)
plt.ylabel(r'$error$', fontsize=16)
plt.title('abs trainingError')
plt.show()
print(err.sum(axis=0)/len(err)*1.0)

#画出测试集误差图
pre_y_test_data = reg.predict(test_x_data)

err_test = abs(pre_y_test_data - test_y_data)

plt.plot(np.linspace(1, len(err_test), len(err_test)), err_test)
plt.xlabel(r'$trainingIndex$', fontsize=16)
plt.ylabel(r'$error$', fontsize=16)
plt.title('abs trainingError')
plt.show()
print(err_test.sum(axis=0)/len(err_test)*1.0)
```
